```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)    #This creates a NumPy array x with values from 0 to 49.
y1 = np.random.randint(0, 100,50) # from zero to 99, 50 random integers
y2 = np.random.randint(0, 100,50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1) # first line 
ax.plot(x,y2) #second line
fig.show()
```

```python
#Adding a legend

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
          frameon = True, #add frame around the legend
          fontsize = 12, #change font size
          ncol = 2, #specify number of columns
          shadow = True )
fig.show()
```

```python
#Move legend outside of the plot area
#Specify where to anchor using bbox_to_anchor

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
          bbox_to_anchor =(1, 1))
fig.show()
```

```python
#Annotations with the text() function

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# text() function to specify coordinates and text
ax.text(10, 95, "This value is important!")
fig.show()
```

```python
#Modifying text annotations

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!",
        ha='center',  ##alignment of text
        color = 'red', ##modify font colour
        size = 20) ##modify font size
fig.show()
```

```python
#Positioning text annotations
# we can locate it at a specific location on our axes or the figure itself
# using "transforms"

#Start by making a simple, empty set of axes
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

#position our annotation with reference to the data , the axes,or the figure 
ax.text(1, 5, "Data:(1, 5)", transform=ax.transData )
ax.text(0.5, 0.1,"Axes:(0.5, 0.1)", transform=ax.transAxes )
ax.text(0.2, 0.2,"Figure:(0.2, 0.2)", transform=fig.transFigure )


```

```python
#Annotating with arrows
#We can use the annotate() function to annotate with arrows and text (e.g. to focus at a specific point)

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
            arrowprops=dict(facecolor='black', shrink=0.05))
fig.show()


```

```python
#Modifying arrows
#Use the arrowprops argument to adjust our arrow style and colour
#Try "->","<->","-[","fancy"


fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!',xy=(10, 95), xytext=(20, 94),
            arrowprops = dict(arrowstyle = "wedge",color = "hotpink"))
fig.show()
```

```python
#Axis Labels

#Removing tick marks and labels
#We can remove tick marks or labels from our axis like so:
    #Use locator to remove ticks and labels
    #Use formatter to remove only labels

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
#Limiting tick marks
#We can also define a maximum number of ticks we want on our axis

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) ##only 3 tick marks
```

```python
#Tick mark intervals
#We can set tick marks at a particular interval using MultipleLocator

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))  #multiples of 5
```

```python
#Rotating axis labels
#To rotate axis labels and horizontal alignment, use xticks()

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, ha='right')  #rotates axis 
```

```python
#Activity: Modifying axis title font

font1 = {'family':'serif','color':'indigo'}
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict = font1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# two plots next to each other, so we just have to define multiple axes and their relative positions:


fig,(ax1, ax2) = plt.subplots(ncols=2,nrows=1,figsize=(7, 3))

ax1.scatter(x1,y1)
ax2.bar(x2,y2)
fig.show()
```

```python
#Subplots without a grid arrangement
#arrange the subplots within our figure by using plt.subplot_mosaic()
#Each axes in our subplot_mosaic() function has a ‘labelʼ
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2','ax3']],
                                    figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)



```

```python
#Layouts
# add some very large x axis titles to our previous plot

fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2','ax3']],
                                    figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```

```python
#Layouts
#see how a constrained layout changes our plot:

fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2','ax3']],
                                    figsize=(7, 4),
                                    layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)

```

```python
#Multiple viz on one axes

#multiple plot methods on the same axes

# first make our sample data
x = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))

# now call both bar and plot elements to the same axes (ax)

ax.bar(x, y1,color = "indigo")
ax.plot(x, y2,color = "red")
```

```python
#Add error information

#calculate standard deviation of data
y2_sd = np.std(y2)

#plot linee
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

#Then use errorbar() to add in our error line (coud be other than stand dev.)

ax.errorbar(x, y2,
            yerr = y2_sd,
            fmt = "none")
```

```python
#customizing error bar
#plot linee
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax.errorbar(x, y2,
            yerr = y2_sd,
            fmt = "none",
            ecolor="indigo",
            elinewidth=4,
            capsize=6,
            capthick=4)
```
