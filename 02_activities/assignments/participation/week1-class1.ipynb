```python
#packages download
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#sample data
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)  #min =0, max =100  (exclusing), 50 randon integers
```

```python
#basic scatterplot
#fig and axis(graph)
fig, ax = plt.subplots(figsize=(5, 3)) #makes a figure plot, "5, 3" is the size
ax.scatter(x,y)   #creates a scatterplot
```

```python
#modifying our graph
#keep our figure and axis definition 

fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)    #creates a bar chart instead
```

```python
#make a line plo
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)    # creates a line graph plot
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)   # we dont use x value
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y) 

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
plt.show()
#fig.tight_layout()  # makes it tighter
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y) 

ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)

plt.show()
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)

```

```python
#Remake our basic scatterplot, but this time add color and marker arguments
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x,y,marker='*', # NEW ADDITIONS
color = "indigo") # NEW ADDITIONS
fig.show()
```

```python
#adjust the style and width of the line connecting our points
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',color = 'indigo',linestyle = '--', # NEW ADDITIONS
linewidth = 2) # NEW ADDITIONS
fig.show()
```

```python
#modify colour using named colors

fig, ax = plt.subplots(figsize=(5, 3))
# color being the new addition
ax.plot(x,y,marker='*', color = '#7425b9', linestyle = '--', linewidth = 2)
fig.show()
```

```python
#Modifying data points

fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x,y,
        marker='*',markersize = 12, # NEW ADDITIONS
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359', # NEW ADDITIONS
        markerfacecolor = '#fa9359' ) # NEW ADDIT
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x,y,
        marker='*',
        markersize = 12, # NEW ADDITIONS
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359', # NEW ADDITIONS
        markerfacecolor = 'black' ) # NEW ADDIT

#We can add grid lines to our axis object using .grid()

ax.grid(axis = 'y')
```
